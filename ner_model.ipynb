```python
import spacy
from spacy import displacy
from spacy.tokens import Span
from spacy.training import Example


try:
    rout = spacy.load("en_core_web_lg")
    print("Model loaded successfully!")
except OSError:
    print("Model not found. Please run 'python -m spacy download en_core_web_lg' in your terminal.")

def run_ner(text):
    doc = rout(text)
    print(f"Entities found in the text:\n{'-'*30}")
    for ent in doc.ents:
        print(f"Entity: {ent.text:<20} Type: {ent.label_:<15} Explanation: {spacy.explain(ent.label_)}")
    
    print(f"\nVisualization:\n{'-'*30}")
    displacy.render(doc, style="ent", jupyter=True)
```

```python
import spacy
import PyPDF2
import docx
import os

try:
    rout = spacy.load("en_core_web_lg")
    print("spaCy model loaded successfully.")
except OSError:
    print("Model not found. Please run 'python -m spacy download en_core_web_lg'.")


def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file."""
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() or "" 
    return text

def extract_text_from_docx(docx_path):
    """Extracts text from a DOCX file."""
    doc = docx.Document(docx_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

def document_to_text(file_path):
    """Routes the file to the correct extraction function based on extension."""
    extension = os.path.splitext(file_path)[1].lower()
    if extension == ".pdf":
        return extract_text_from_pdf(file_path)
    elif extension == ".docx":
        return extract_text_from_docx(file_path)
    elif extension == ".doc":
        return "Legacy .doc format extraction not supported in this simple script. Use .docx or PDF."
    else:
        return None


def process_document_with_ner(file_path):
    """Extracts text from a file and runs the NER model."""
    print(f"Processing file: {file_path}")
    raw_text = document_to_text(file_path)

    if raw_text:
        doc = rout(raw_text)
        print(f"\n--- NER Results for {os.path.basename(file_path)} ---")
        for ent in doc.ents:
            print(f"Entity: {ent.text:<20} Type: {ent.label_}")
        
        

    else:
        print(f"Could not extract text from {file_path}. Format might not be supported.")

```

```python
process_document_with_ner("sample.pdf")
print("-" * 50)
process_document_with_ner("sample.docx")
```

```python
import spacy
from spacy.training import Example
from spacy.scorer import Scorer
from spacy.tokens import Doc
import os
import PyPDF2
import docx

try:
    rout = spacy.load("en_core_web_lg") 
    print("spaCy model 'en_core_web_lg' loaded successfully.")
except OSError:
    print("Large model not found. Loading fallback 'en_core_web_sm'...")
    rout = spacy.load("en_core_web_sm")  
    print("spaCy model 'en_core_web_sm' loaded instead.")


def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() or ""
    return text

def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    full_text = [para.text for para in doc.paragraphs]
    return '\n'.join(full_text)



PDF_GOLD_DATA = [
    (extract_text_from_pdf(r"C:\Users\deepa\Cloud_Nexus\Cloud_Nexus\ner_plain_dataset.pdf"),  
     {"entities": [
         (58, 64, "ORG"),      
         (69, 78, "ORG"),      
         (80, 88, "PER"),      
         (107, 116, "ORG"),    
         (121, 134, "PER"),   
         (162, 167, "GPE"),   
         (183, 194, "DATE"),   
         (220, 228, "GPE"),    
         (262, 273, "DATE"),   
     ]}
    )
]

DOCX_GOLD_DATA = [
    (extract_text_from_docx(r"C:\Users\deepa\Cloud_Nexus\Cloud_Nexus\ner_plain_dataset.docx"),   
     {"entities": [
         (106, 127, "ORG"),    
         (146, 152, "GPE"),    
         (177, 186, "MONEY"),  
         (206, 214, "PER")     
     ]}
    )
]

# Combined dataset
TEST_DATA = PDF_GOLD_DATA + DOCX_GOLD_DATA


# --- 3. Evaluation Function ---

def evaluate_model(model, test_data):
    scorer = Scorer()
    examples = []
    for text, annotations in test_data:
        doc = model.make_doc(text)
        example = Example.from_dict(doc, annotations)
        example.predicted = model(text)  # Run prediction
        examples.append(example)

    scores = scorer.score(examples)

    print("\n" + "="*60)
    print("NER Model Evaluation Results (using sample PDF/DOCX text):")
    print("="*60)
    print(f"Overall Precision: {scores['ents_p'] * 100:.2f}%")
    print(f"Overall Recall:    {scores['ents_r'] * 100:.2f}%")
    print(f"Overall F1-Score:  {scores['ents_f'] * 100:.2f}%")
    print(f"Token Accuracy:    {scores.get('token_acc', 0) * 100:.2f}%")
    print("="*60)


# --- 4. Run Evaluation ---
evaluate_model(rout, TEST_DATA)
```
